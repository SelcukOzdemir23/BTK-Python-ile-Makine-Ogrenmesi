```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Churn_Modelling.csv")
```

```python
df.head()
```

```python
dummies_geography = pd.get_dummies(df["Geography"])
dummies_gender= pd.get_dummies(df["Gender"])
```

```python
dummies_gender
```

```python
df.drop(["Geography","Gender","RowNumber","CustomerId","Surname"],axis=1,inplace=True)
df
```

```python
df = pd.concat([df,dummies_gender,dummies_geography],axis=1)
df
```

```python
X = df[df[df.columns]!="Exited"]
Y = df["Exited"]
```

```python
print(Y)
X
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.33,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc =StandardScaler()
```

```python
X_train
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

# Yapay Sinir AÄŸlarÄ±

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
classifier.add(Dense(6,activation="relu",input_dim=14)) # giriÅŸ katmanÄ±
classifier.add(Dense(6,activation="relu")) # orta katman
classifier.add(Dense(1,activation="sigmoid")) #son katman- Ã§Ä±kÄ±ÅŸ
```

```python
classifier.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"],
    )
```

```python
classifier.fit(X_train,Y_train,epochs=7)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred = (y_pred >0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,y_pred)
cm
```
