```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Churn_Modelling.csv")
```

```python
df.head()
```

```python
dummies_geography = pd.get_dummies(df["Geography"])
dummies_gender= pd.get_dummies(df["Gender"])
```

```python
dummies_gender
```

```python
df.drop(["Geography","Gender","RowNumber","CustomerId","Surname"],axis=1,inplace=True)
df
```

```python
df = pd.concat([df,dummies_gender,dummies_geography],axis=1)
df
```

```python
X = df[df[df.columns]!="Exited"]
Y = df["Exited"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.33,random_state=0)
```

# XGBOOST

```python
from xgboost import XGBClassifier
```

```python
classifier = XGBClassifier()
classifier.fit(X_train,Y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_pred,Y_test)
cm
```
