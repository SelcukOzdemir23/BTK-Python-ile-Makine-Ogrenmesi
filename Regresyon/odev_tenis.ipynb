```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
```

```python
veriler = pd.read_csv("odev_tenis.csv")
```

```python
veriler = pd.concat([veriler,pd.get_dummies(veriler["outlook"])],axis=1)
veriler.drop(["outlook"],axis=1,inplace=True)
```

```python
veriler
```

```python
windy = []
for i in veriler["windy"].values:
    windy.append(int(i))
```

```python
windy
```

```python
veriler["windy_cat"] = windy
```

```python
veriler.drop(["windy"],axis=1,inplace=True)
```

```python
veriler
```

```python
veriler["play"].replace(["yes","no"],[1,0],inplace=True)
```

```python
veriler
```

```python
X = veriler.drop(["temperature"],axis=1)
Y= veriler["temperature"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =train_test_split(X,Y,random_state=0,test_size=0.33)
```

```python
linearModel = LinearRegression()
```

```python
linearModel.fit(x_train,y_train)
```

```python
linearModel.score(x_test,y_test)
```

```python
linearModel.predict(x_test)
```
