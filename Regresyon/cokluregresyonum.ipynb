# Eksik Veriler

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("veriler.csv")
```

```python
df
```

```python
c = df.iloc[:,-1:].values
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
c[:,-1] = le.fit_transform(df.iloc[:,-1])
ohe = preprocessing.OneHotEncoder()
c = ohe.fit_transform(c).toarray()
```

```python
sonuc3 = pd.DataFrame(data=c[:,:1],index=range(22),columns=["cinsiyeti"])
```

```python
df = pd.concat([df,sonuc3],axis=1)
```

```python
df.drop(["cinsiyet"],axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
df= pd.concat([df,pd.get_dummies(df["ulke"])],axis=1)
df.drop(["ulke"],axis=1,inplace=True)
X = df.drop(["boy"],axis=1)
Y = df["boy"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.33,random_state=0)
x_test
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_predict = regressor.predict(x_test)
```

```python
regressor.score(x_test,y_test)
```
